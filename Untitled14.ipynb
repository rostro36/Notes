```python
train_center = tf.keras.utils.image_dataset_from_directory(
  'drive/MyDrive/AID/AID/Center/',
  validation_split=0.2,
  subset="training",
  color_mode="grayscale",
  seed=123,
  labels=None,
  label_mode=None,
  image_size=(600, 600),
  batch_size=32)
train_church = tf.keras.utils.image_dataset_from_directory(
  'drive/MyDrive/AID/AID/Church',
  validation_split=0.2,
  subset="training",
  color_mode="grayscale",
  seed=123,
  labels=None,
  label_mode=None,
  image_size=(600, 600),
  batch_size=32)
val_center = tf.keras.utils.image_dataset_from_directory(
  'drive/MyDrive/AID/AID/Center',
  validation_split=0.2,
  subset="validation",
  color_mode="grayscale",
  seed=123,
  labels=None,
  label_mode=None,
  image_size=(600, 600),
  batch_size=32)
val_church = tf.keras.utils.image_dataset_from_directory(
  'drive/MyDrive/AID/AID/Church',
  validation_split=0.2,
  subset="validation",
  color_mode="grayscale",
  seed=123,
  labels=None,
  label_mode=None,
  image_size=(600, 600),
  batch_size=32)
test_ds = train_center.take(5).concatenate(train_church.take(5))
training_ds = train_center.skip(5).concatenate(train_church.skip(5))
training_ds = training_ds.shuffle(buffer_size=len(training_ds), seed=32)
validation_ds = val_center.concatenate(val_church)
validation_ds = training_ds.shuffle(buffer_size=len(validation_ds), seed=42)
```

```python
# https://www.tensorflow.org/tutorials/generative/autoencoder


class AutoEncoder(Model):
  def __init__(self, filters=16,window=5, dropout=0.1, encoding_dimension=128):
    super(AutoEncoder, self).__init__()
    self.encoder = tf.keras.Sequential([
      layers.Input(shape=(600, 600, 1)),
      layers.Conv2D(filters, (window, window), activation='selu', padding='same', strides=(int( (window-1)/2),int((window-1)/2)) ),
      layers.Conv2D(int(filters/2), (window, window), activation='selu', padding='same', strides=(int((window-1)/2),int((window-1)/2))) ])
    self.encoder.add(layers.Flatten(name="flatten_encoder"))
    flatten_layer = self.encoder.get_layer('flatten_encoder')
    print(flatten_layer.output_shape[1])
    print(flatten_layer.input_shape)
    self.encoder.add(layers.Dense((flatten_layer.output_shape[1]+encoding_dimension)/2, activation="selu"))
    self.encoder.add(layers.Dropout(dropout))
    self.encoder.add(layers.Dense(encoding_dimension, activation="selu"))

    self.decoder = tf.keras.Sequential([
      layers.Dense((flatten_layer.output_shape[1]+encoding_dimension)/2, activation="selu"),
      layers.Dropout(dropout),
      layers.Dense(flatten_layer.output_shape.output_shape, activation="selu"),
      layers.Reshape(flatten_layer.output_shape.input_shape),
      layers.Conv2DTranspose(filters/2, kernel_size=window, strides=( (window-1)/2, (window-1)/2), activation='selu', padding='same'),
      layers.Conv2DTranspose(filters, kernel_size=window, strides=( (window-1)/2, (window-1)/2), activation='selu', padding='same'),
      layers.Conv2D(1, kernel_size=(window, window), activation='selu', padding='same')])

  def call(self, x):
    encoded = self.encoder(x)
    decoded = self.decoder(encoded)
    return decoded

autoencoder = AutoEncoder()
default_lr = 0.001,
optimizer = tf.keras.optimizers.Adam(default_lr)
autoencoder.compile(optimizer=optimizer, loss=losses.MeanSquaredError())
autoencoder.summary()
```
